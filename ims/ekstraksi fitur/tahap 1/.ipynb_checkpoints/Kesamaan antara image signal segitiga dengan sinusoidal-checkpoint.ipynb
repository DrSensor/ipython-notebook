## Shi-Tomasi Corner Detector

```python
%matplotlib inline

import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('triangle.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

height, width = img.shape[:2]
result = np.ones((height, width, 3), np.uint8)
result[:] = tuple(reversed((255,255,255)))
```

```python
plt.imshow(img)
plt.show()
```

```python
corners = cv2.goodFeaturesToTrack(gray,25,0.01,10)
corners = np.int0(corners)
corners
```

```python
%matplotlib inline

for i in corners:
    x,y = i.ravel()
    cv2.circle(result,(x,y),10*(width/height),255,-1)
plt.imshow(result)
plt.show()
```

```python
img &= result
plt.imshow(img)
plt.show()
```

## Liniear Regression

```python
from numpy.polynomial import Polynomial

xi =[]
yi = []

for i in corners:
    x,y = i.ravel()
    xi += [x]
    yi += [y]
    
m,c = np.polyfit(xi, yi, 1)
print 'y =',c,'+',str(m)+'x'
y = lambda x: int(c+m*x)
# pts = *p.linspace()
# plt.plot(xi,yi)
# plt.plot(*p.linspace())
# plt.show()
```

```python
y = lambda x: int(c+m*x)
result = cv2.line(result, (0,y(0)), (width, y(width)), color=(0,255,0), thickness=10*(width/height))
# result = cv2.polylines(result, p.linspace()[0], False, color=(0,255,0), thickness=10*(width/height))
print (xi[0],y(0)), (xi[len(xi)-1], height)
plt.imshow(result)
```

```python
img &= result
plt.imshow(img)
plt.show()
```

## Slope dari Titik-titik puncak

http://www.madematika.com/2014/10/menentukan-persamaan-garis-lurus_17.html

```python
from sympy import Point, Line

L = Line(Point(0,y(0)), slope=m)
Ls = []
peak_line = []
for i in corners:
    xs,ys = i.ravel()
    l = L.perpendicular_segment(Point(xs,ys))
    peak_line += [(float(l.slope),float(l.coefficients[2]))]
    Ls += [l]

for p in peak_line: 
    print 'y =',p[0],'+',str(p[1])+'x'
# print float(L.coefficients[0])
# print L.equation()
```
